```python
import pandas as pd
import requests
```

запрос данных

```python
visits = requests.get('https://data-charts-api.hexlet.app/visits?begin=2022-06-01&end=2023-06-01', auth=('user', 'pass'))
data_visits = visits.json()
df_visits = pd.json_normalize(data_visits)
df_visits
```

```python
regs = requests.get('https://data-charts-api.hexlet.app/registrations?begin=2022-06-01&end=2023-06-02', auth=('user', 'pass'))
data_regs = regs.json()
df_regs = pd.json_normalize(data_regs)
df_regs
```
